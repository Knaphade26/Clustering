```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
```

# Hierarchical

```python
crime = pd.read_csv('D:\ExcelR\Assignment\Clustering\crime_data.csv')
crime
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(crime.iloc[:,1:])
df_norm
```

```python
dendrogram = sch.dendrogram(sch.linkage(df_norm, method = "average"))
```

```python
hc=AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage = 'complete')
```

```python
y_hc = hc.fit_predict(df_norm)
y_hc
```

```python
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
crime['h_clusterid'] = Clusters
crime
```

# k-means

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_crime= scaler.fit_transform(crime.iloc[:,1:])
```

```python
scaled_crime
```

```python
wcss = []
for i in range(1, 21):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_crime)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 21), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
cluster_new = KMeans(4, random_state=42)
cluster_new.fit(scaled_crime)
```

```python
cluster_new.labels_
```

```python
crime['clusteried_new']=cluster_new.labels_
```

```python
crime
```
